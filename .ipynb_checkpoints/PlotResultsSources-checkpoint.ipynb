# EFTCAMB-sources results

Here we want to show the results obtained for the cross correlations between CMB and galaxy-counts

The cross correlations are given by 

$$ C_{\ell} \propto \int dk \, k^2 \Delta_{\ell}^{\rm ISW}(k) \Delta_{\ell}^{g}(k)\mathcal{P}_{\mathcal{R}}(k), $$

where the transfer functions are given by

$$ \Delta_{\ell}^{\rm ISW}(k) = -\int_{\eta_*}^{\eta_0} d \eta \, (\Phi^{\prime} + \Psi^{\prime}) \, j_{\ell} [k (\eta_0-\eta)], $$

$$ \Delta_{\ell}^{g}(k) = \int_{0}^{\eta_0} d \eta \, W(\eta) \, b_g (\eta,k) \,  \delta(\eta, k) \, j_{\ell}[k(\eta_0 -\eta)]$$

and the primordial power spectrum  is $\mathcal{P}_{\mathcal{R}}(k) = A_s \left( \frac{k}{k_0}\right)^{n_s-1}$ with the pivot scale $k_0$

## Statistical errors

Here is just an example of the statistical errors as in https://arxiv.org/abs/astro-ph/0606626

The signal to noise ratio at each multipole $\ell$ is given by

$$ \left( \frac{S}{N}\right)_{\ell}^2 = \frac{f_{\rm sky} (2\ell+1)}{4 \pi} \frac{ \left( C_{\ell}^{Tg} \right)^2}{ C_{\ell}^{TT} C_{\ell}^{gg} + \left( C_{\ell}^{Tg} \right)^2} $$

and the statistical relative error at each multipole is given by $(S/N)^{-1}$.

The theoretical statistical error on each bin is hence given by 

$$\Delta C_{\ell}^{Tg} = \left( \frac{S}{N} \right)^{-1}_{\ell} C_{\ell}^{Tg} $$

```python
# fraction of the sky covered by the survey
f_sky = 1.0
```

```python
LCDM_W1xW1 = LCDM_df.W1xW1.values
LCDM_W2xW2 = LCDM_df.W2xW2.values
LCDM_W3xW3 = LCDM_df.W3xW3.values
Cl11 = 2*3.1415/ell/(ell+1)*LCDM_W1xW1
Cl11 = 2*3.1415/ell/(ell+1)*LCDM_W1xW1
Cl11 = 2*3.1415/ell/(ell+1)*LCDM_W1xW1
```

```python
SNr_LCDM_W1 = f_sky*(2*ell+1)/(4*3.1415)*((LCDM_TW1**2)/(LCDM_TT* LCDM_W1xW1 + LCDM_TW1**2))
SNr_LCDM_W2 = f_sky*(2*ell+1)/(4*3.1415)*((LCDM_TW2**2)/(LCDM_TT* LCDM_W2xW2 + LCDM_TW2**2))
SNr_LCDM_W3 = f_sky*(2*ell+1)/(4*3.1415)*((LCDM_TW3**2)/(LCDM_TT* LCDM_W3xW3 + LCDM_TW3**2))
# total signal to noise
SN1 = SNr_LCDM_W1.sum()
SN2 = SNr_LCDM_W2.sum()
SN3 = SNr_LCDM_W3.sum()
print('1: S/N =', SN1)
print('2: S/N =', SN2)
print('3: S/N =', SN3)
```

```python
plt.semilogx(ell, SNr_LCDM_W1)
plt.semilogx(ell, SNr_LCDM_W2)
plt.semilogx(ell, SNr_LCDM_W3)
plt.show()
```

The statistical error in each bin of multipoles are given by

$$ \Delta C_{\bar{\ell} \,{\rm bin}} = \left[ \sum_{\ell \in {\rm bin}} \left( \frac{S}{N} \right)_{\ell}^{2} \right]^{-1} $$

This has to be corrected..

```python
# defining the bins
# 1: 2-18
SNr_bin = np.zeros(4)
#SNr_bin[0] = np.sum(np.sqrt(SNr_LCDM_W3[0:15]**(-2)))
SNr_bin[0] = (np.sum(np.sqrt(SNr_LCDM_W1[0:19]**2)/SNr_LCDM_W1[0:19]))**(-1)
#SNr_bin[0] = 
# 2: 19-49
SNr_bin[1] = (np.sum(np.sqrt(SNr_LCDM_W1[20:49]**2)/SNr_LCDM_W1[20:49]))**(-1)
# 3: 50-99
SNr_bin[2] = (np.sum(np.sqrt(SNr_LCDM_W1[50:99]**2)/SNr_LCDM_W1[50:99]))**(-1)
# 4: 100-800
SNr_bin[3] = (np.sum(np.sqrt(SNr_LCDM_W1[100:]**2)/SNr_LCDM_W1[100:]))**(-1)
```

```python
ell_bin = np.zeros(4)
ell_bin[0]= ell[10]
ell_bin[1]= ell[35]
ell_bin[2]= ell[75]
ell_bin[3]= ell[450]
```

```python
# central values of the cross correlation
CCF_bin = np.zeros(4)
CCF_bin[0] = LCDM_TW1[10]
CCF_bin[1] = LCDM_TW1[35]
CCF_bin[2] = LCDM_TW1[75]
CCF_bin[3] = LCDM_TW1[450]
```

```python
#now plotting
plt.semilogx(ell,LCDM_TW1)
plt.errorbar(ell_bin, CCF_bin,  yerr=SNr_bin ,fmt='o', markersize=8)
plt.xlim(2,800)
pil.savefig('ISW_errorbars.pdf', bbox_inches='tight')
plt.show()
```

## Triangular Plot for Galaxy Number Counts and CMB temperature cross correlations

Here I want to combine in a single plot the auto- and cross-correlations of the CMB temperature and Galaxy Number Counts (GNC) power spectra. The GCN power spectra are defined by

$$ C_{\ell} \propto \int \frac{dk}{k} \Delta_{\ell}^{W_i}(k) \Delta_{\ell}^{W_j}(k)\mathcal{P}_{\mathcal{R}}(k), $$

where 

$$ \Delta_{\ell}^{W_i}(k) = \int_{0}^{\eta_0} d \eta \, W^{i}(\eta) \, b_g (\eta,k) \,  \delta(\eta, k) \, j_{\ell}[k(\eta_0 -\eta)], \quad \quad i =1,2,3. $$
